--> how to write a class

```python
class Rectangle:
    def __init__(self):
        self.length=10 # properties are defined in init method 
        # which take the parameter self
        #property are called field
        self.breadth=5
       
    def area(self): #every funtion should have para meter self
        return self.length*self.breadth
    
    def perimeter(self):
        return 2*(self.length+self.breadth)
    
r=Rectangle()
print('length',r.length)
print('breadth',r.breadth)
print('area',r.area())
print('perimeter',r.perimeter())
```

CIRCLE CLASS

```python
import math

class Circle:
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return math.pi*self.radius**2
    
    def perimeter(self):
        return 2*math.pi*self.radius
    
a=Circle(5)
print(f"Area:{a.area():.2f}")
print(f"perimeter:{a.perimeter():.2f}")
```

 CLASS FOR BANK ACCOUNT
----- NAME      
----- ACC NO(should be given by the bank)
----- BALANCE

```python
class InsufficientFundsError(Exception):
    pass
class BankAccount:
    acc_counter=100

    def __init__(self,name,balance):
        BankAccount.acc_counter+=1
        self.name=name
        self.acc=BankAccount.acc_counter
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount

    def withdraw(self,amount):
        if 0<amount<=self.balance:
            self.balance-=amount
            return f"withdraw {amount}.new_balance={self.balance}"
        else:
            raise InsufficientFundsError("not enough funds")
        
    def get_balance(self):
        return self.balance
    
    def details(self):
        return self.acc,self.name,self.balance
    
james=BankAccount('James',1000)
ashok=BankAccount('ashok',1000)
print(james.withdraw(500))
print(james.get_balance())
print(ashok.details())
```
