## INHERITENCE

WHEN ONE CLASS INHERITES PROPERTY FROM THE PREVIOUS CLASS IT IS CALLED 
INHERITENCE

```python
#CUBOID ---->INHERITS LENGTH,BERADTH FROM RECTANGLE BUT FOR HEIGHT 
# WE HAVE TO MAKE NEW CLASS

class Rectangle:
    def __init__(self,A,B):
        self.length=A
        self.breadth=B
       
    def area(self): #every funtion should have para meter self
        return self.length*self.breadth
    
    def perimeter(self):
        return 2*(self.length+self.breadth)

class Cuboid(Rectangle):

    def __init__(self,h):
        self.height=h

    def volume(self):
        return self.length*self.breadth*self.height
    
c=Cuboid(3)
print(c.volume()) #
```

the above one is wrong because length and breadth is not defined in cuboid class

```python
class Rectangle:
    def __init__(self,A,B):
        self.length=A
        self.breadth=B
       
    def area(self): #every funtion should have para meter self
        return self.length*self.breadth
    
    def perimeter(self):
        return 2*(self.length+self.breadth)

class Cuboid(Rectangle):

    def __init__(self,A,B,h):
        self.height=h
        super().__init__(A,B)

    def volume(self):
        return self.length*self.breadth*self.height
    
c=Cuboid(10,5,3)
print(c.volume()) #
```

Shape Area Calculation--->

```python
import math
class Shapes:
    def area(self):
        print("Area not defined")

#subclass--> Circle
class Circle(Shapes):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        area_value= math.pi*(self.radius**2)
        print(f"area of circle:{area_value:.2f}")

#subclass---> Rectangle

class Rectangle(Shapes):
    def __init__(self,length,bridth):
        self.length=length
        self.bridth=bridth
    def area(self):
        area_value=self.length*self.bridth
        print(f"Area of the rectangle:{area_value}")

c=Circle(5)
c.area()

r=Rectangle(3,4)
r.area()

```

BANK ACCOUNT SYSTEM

```python
class Account:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
 
    def deposit(self,amount):
        self.balance+=amount
        print(f"Your balance after money deposit is : {self.balance}")
   
    def withdraw(self,w_amount):
        if w_amount<=self.balance:
            self.balance-=w_amount
            print(f"After withdrawal your balance : {self.balance}")
        else :
            print("Insufficient balance")
#subclass
    def show_balance(self):
        print(f"account holder :{self.name}|Balance :{self.balance}")

class SavingAccount(Account):
    def __init__(self,name,balance,interest_rate):
        super().__init__(name,balance)
        self.interest_rate=interest_rate

    def add_interest(self):
        interest=self.balance*self.interest_rate
        self.balance+=interest
        print(f"after adding interest: {self.balance}")

        
s_acc=SavingAccount("harshita",5000,0.05)
print("\n--saving account----")
s_acc.show_balance()
s_acc.deposit(1000)
s_acc.add_interest()
s_acc.withdraw(2000)
    
       
```
