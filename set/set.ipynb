```python
# 14. Find Maximum and Minimum Values in a Set

# Write a Python program to find the maximum and minimum values in a set.
a={5,10,15,20,2}

# max_val=max(a)
# min_val=min(a)
# print(max_val)
# print(min_val)

              #or

a={5,10,15,20,2}

if not a:
    print("set is empty")
else:
    iterator=iter(a)
    max_val=min_val=next(iterator)

    for i in iterator:
        if i>max_val:
            max_val=i
        if i<min_val:
            min_val=i
print(max_val)
print(min_val)
```
