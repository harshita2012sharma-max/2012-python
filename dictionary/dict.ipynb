```python
#question- we have to find keys with max value
scores={"a":50,"b":80,"c":80,"d":60}
Max=0
max_i=None
for i,v in scores.items():
    if v>Max:
        Max=v
        max_i=[i]
    elif v==Max:
        max_i.append(i)
print(max_i)   
print(Max)
```
