```python
#dict comprehension-------->
#1. similar to list comprehension they allow you to construct dict in a single line with an optional condition for filtering 
#2.syntax:
# (key_expression:value_expression for item in iterable if condition)
```

```python
#question-1 
squares_dict={x:x**2 for x in range(1,8)}
squares_dict
```

```python
#question-2
even_dict={x:x**2 for x in range(1,11) if x%2==0}
even_dict
```

```python
dict={x:(x**2 if  x%2==0 else x )for x in range(10)}
dict
```

```python
#question-3 reverse a dictionary
original={"a":1,"b":2,"c":3}
reverse={v:k for k,v in original.items()}
reverse
```

```python
#question-4
words=["apple","banana","cherry"]
length_dict={word:len(word) for word in words}
length_dict
```

```python
#question-create a dict where keys are words and values are in uppercase
words=["python","code","dict"]
dictionary={x:x.upper()  for x in words}
dictionary
```

```python
#question-create a dictionary from a sentence with each words and its frequency.
sentence="python makes coding and python is fun"
sentence=sentence.split()
d={word:sentence.count(word) for word in sentence}
d
```

```python
#question- creare a nested dict of numbers from 1-3 with their  squares and cube.
d={x:{'squares':x**2,'cube':x**3} for x in range(1,4)}
d
```

```python
#question-create a dict froma string where keys are vowels and value are count
text="dictionary comprehension practice"
text=text.lower()
d={x:text.count(x) for x in text if x in "aeiou"}
d
```

```python
#question-create a dict with 1-10 as keys and "divisible by 3"/ "not divisible " as values
d={x:"divisible by 3" if x%3==0 else "not divisible"for x in range(1,10)}
d
```

```python
#question-create a dict from 2 lists.
keys=['name','age','city']
values=['sam',30,'delhi']
d={}
for i in range(len(keys)):
    d[keys[i]]=values[i]
d

d={i:j for i,j in zip(keys,values)}
d
```
