#q-4 write a program to input a string and print the freq of each vowel using only a for loop (no dict)

```python
s = input("Enter a string: ").lower()  
a_count = e_count = i_count = o_count = u_count = 0


for ch in s:
    if ch == 'a':
        a_count += 1
    elif ch == 'e':
        e_count += 1
    elif ch == 'i':
        i_count += 1
    elif ch == 'o':
        o_count += 1
    elif ch == 'u':
        u_count += 1
print("Frequency of vowels:")
print("a:", a_count)
print("e:", e_count)
print("i:", i_count)
print("o:", o_count)
print("u:", u_count)
```

```python
#q-1 perfect number
# Function to check perfect number

def is_perfect(num):
    sum_div = 0
    for i in range(1, num):
        if num % i == 0:
            sum_div += i
    return sum_div == num

print("Perfect numbers between 1 and 1000:")

for n in range(1, 1000):
    if is_perfect(n):
        print(n)
```
