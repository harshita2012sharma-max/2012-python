```python
class Person:
    def __init__(self,name,id,gender):
        self.name=name
        self.id=id
        self.gender=gender

    def display(self):
        print(f"name:{self.name}")
        print(f"id:{self.id}")
        print(f"gender:{self.gender}")


class Student(Person):   #-------->inheriting all the attribute from the parents class to the child class
    #super().__init__(self,name,id,gender)
    def __init__(self, name, id, gender,marks):
        super().__init__(name, id, gender)
        self.marks=marks
    
    def display(self):
        super().display()   #accessing the parents class
        print(f"marks:{self.marks}")

class Teacher(Person):
    def __init__(self, name, id, gender,salary,department):
        super().__init__(name, id, gender)
        self.salary=salary
        self.department=department

    def display(self):
        super().display()
        print(f"salary:{self.salary}")
        print(f"department:{self.department}")
s1=Student("rahul",10,"male",40)
t1=Teacher("Shyam",102,"Male",20000,"CSE")
print("teacher info")
t1.display()
print("student info")
s1.display()
```

```python
class Animal:
    def __init__(self, type, gender):
        self.type = type
        self.gender = gender
    
    def display(self):
        print(f"Type: {self.type}")
        print(f"Gender: {self.gender}")

class Category(Animal):
    def __init__(self, type, gender, breed, sound):
        super().__init__(type, gender)
        self.breed = breed
        self.sound = sound
    
    def display(self):
        super().display()
        print(f"Breed: {self.breed}")
        print(f"Sound: {self.sound}")

# Create Category objects (since they have breed & sound)
a1 = Category("Cat", "Female", "Persian", "Meow")
a2 = Category("Dog", "Male", "Golden Retriever", "Bark")

a1.display()
print()
a2.display()
```

```python
class Vehical:
    def __init__(self,brand,color):
        self.brand=brand
        self.color=color

    def display(self):
        print(f"brand:{self.brand}")
        print(f"color:{self.color}")

class Car(Vehical):
    def __init__(self,brand,color,speed,price):
        super().__init__(brand,color)
        self.speed=speed
        self.price=price
    
    def display(self):
        super().display()
        print(f"speed:{self.speed}")
        print(f"price:{self.price}")

c1=Car("BMW","black",220,500000)

c1.display()
```

```python
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    
    def display(self):
        print(f"name:{self.name}")
        print(f"age : {self.age}")
        print(f"gender : {self.gender}")

class Singer(Person):
    def __init__(self,name,age,gender,no_song,genre,album_released):
        super().__init__(name,age,gender)
        self.no_song=no_song
        self.genre=genre
        self.album_released=album_released

    def display(self):
        super().display()
        print(f"no_song:{self.no_song}")
        print(f"genre:{self.genre}")
        print(f"album_released:{self.album_released}")
 
class Teacher(Person):
    def __init__(self,name,age,gender,yr_exp,qualification):
        super().__init__(name,age,gender)
        self.yr_exp=yr_exp
        self.qualification=qualification

    def display(self):
        super().display()
        print(f"yr_exp:{self.yr_exp}")
        print(f"qualifiaction:{self.qualification}")


s1=Singer("harshita",20,'female',24,'jazz','young forever')
t1=Teacher("jklm",24,'male',3,"everything")
s1.display()
print("--------------------------")
t1.display()
        
```

```python
class Bank:
    def __init__(self,name,balance):
        self.name=name
        self.__balance=balance

    #getter
    def display_bal(self):
        print(f"you balance is :::{self.__balance}")

    #setter
    def withdraw(self,amount):
        if amount<self.__balance:
           self.__balance-=amount
           print(f"Now your remaning balance is :::{self.__balance}")
        else:
            print("your amount is more than the balance")
    def deposit(self,amount):
        self.__balance+=amount
        print(f"After deposit your current balance is ::: {self.__balance}")

b=Bank("harshita",3000000000)
print("--------BANK MANAGEMENT------------")
b.display_bal()
b.withdraw(100000000)
b.deposit(200000000)
```
