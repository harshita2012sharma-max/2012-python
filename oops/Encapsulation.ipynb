Encapsulation---->bundles data and the methods that operate on the data into a single unit,like a class and retricts direct access to the object's internal state.
#private
#protected
#public    ------<|> access attributes

```python
class Bankacc:
    def __init__(self,acc_no,balance):
        self.acc_no=acc_no
        self.balance=balance
acc=Bankacc(1234,50000)
acc.balance=-2000
print(acc.balance)
```

```python
#question-- 
class student:
    def __init__(self,name,marks): ## if we wnat to make our attribute private then we use __
        self.name=name
        self.__marks=marks
    def show(self):
        return self.__marks   #getter
    
    def change(self,marks):  #setter
        self.__marks=marks
    
s1=student("rahul",50)
s1.change(60)
print(s1.name)
print(s1.show())

```

```python
class Bank_Account:
    def __init__(self,account_holder="harshita",balance=100):
        self.account_holder=account_holder
        self.__balance=balance
    def show(self):
        return self.__balance
    def change(self,balance):
        self.__balance=balance

acc1=Bank_Account()
acc1.change(200000000000)
print(acc1.account_holder)
print(acc1.show())
```

```python
class product:
    def __init__(self,name,price,quantity):
        self.__name=name
        self.__price=price
        self.__quantity=quantity
    def show(self):
        return self.__name,self.__price,self.__quantity
    def change(self,name,price,quantity):
        self.__name=name
        self.__price=price
        self.__quantity=quantity
    def total_value(self):
        return self.__price*self.__quantity
p1=product("diamond",2345678910,2)
p1.change("jklm",4,3)

print(p1.show())
print(p1.total_value())
```
