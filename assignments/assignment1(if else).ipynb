```python
#question-6
""" a year is leap year if 
1. It is divided by 4 or 400.
2. It is not divided by 100.
"""
year=int(input("enter the year : "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print("it is a leap year.")
else:
    print("it is not a leap year.")
    #OR
if year%4==0:
    if (year%100!=0 or year%400==0):
        print("it is a leap year.")
    else:
        print("it is not a leap year.")     
else:
    print("it is not a leap year.")   
```

```python
#question-11
first_side=float(input("enter the first side:"))
second_side=float(input("enter the second side:"))
third_side=float(input("enter the third side:"))

if first_side==second_side==third_side:
    print("equilateral triangle")
elif (first_side==second_side!=third_side) or (first_side==third_side!=second_side) or (second_side==third_side!=first_side):
    print("isosceles triangle ")
elif first_side!=second_side!=third_side:
    print("scalene triangle")      
```
