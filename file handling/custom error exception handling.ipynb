```python
def age_validate(age):          # raise ---->keyword that throws exception if error found
    if age<0:                    # exception--->it is general purpose class
        raise Exception("please enter the valid age")
    else:
        print("age is validate")

try:
    age_validate(-1)
except Exception as E:
    print("you have got this exception :",E)
```

```python
#Bank management
amount=10000
def cash_withdraw(cash):
    global amount
    if int(cash)>int(amount):
        raise Exception("your withdrawal has cross the amount limit")
    else:
        amount=amount-cash
        print("your transaction is completed")
        print("rest money",amount)
try:
    cash_withdraw(2000)
except Exception as E:
    print("Warning:",E)

```

```python
def enter_marks(marks):
    if marks<0 or marks>100:
        raise Exception("the given marks are not valid")
    else:
        print(f"your total marks are:{marks}")

try:
    enter_marks(-1)
except Exception as E:
    print("warning:", E)
```

```python
def enter_password(password):
    if len(password)<8:
        raise Exception("Your given password is weak")
    else:
        print("Strong Password")
try:
    enter_password('harshita2005')
except Exception as E:
    print("warning:", E)    
```

```python
def email_signup(email):
    if '@' not in email:
        raise Exception("wrong email")
    else:
        print("right email")
try:
    email_signup('harshita2005')
except Exception as E:
    print("warning:", E)    
```
