```python
# why do we meed exception handling
# when we write code sometimes unexpected sitation occur
#if python encounters an error and there is no error handling the progrm crashes
#beacuse user dont wang the crashes
```

```python
#zero division error
a=10
b=0
# this try blok always runs
try :
    c=a/b
#this block runs when any exception error occurs into program
except ZeroDivisionError:
    print("plese enter the value grater than 0")
#this block will only run when any exception doesn't occure
else:
    print(c)
```

```python
try:
    print(z)
except NameError:
    print("please define the variable which you are using")
```

```python
#type error
try:
    a='str'
    b=10
    c=a+b
except TypeError:
    print("use same data type fro this operation ")
```
