```python
#slicing in list
a=["sam",45,56.90,"rahul",[12,"raj",78.56]]
print(a[0:2])
print(a[1:4])
print(a[2:])
print(a[:])
print(a[-1::-1])
print(a[3][2])
print(a[4][1][1])
```

```python
#question-create a user defined list and then you have to  do max of numbers

a=[]
size=int(input("enter the size : "))
for i in range(size):
    val=int(input("enter the numbers"))
    a.append(val)
print(a)
max=a[0]
for i in range(len(a)):
    if a[i] > max:
        max=a[i]
print(max)
```
