```python
#q-check if number is palindrom
def m_rev(num):
    
    original=num
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num=num//10
    print(rev)
    if rev==original:
      print("palindrom")
    else:
      print("not")
n=int(input("enter the number : "))
m_rev(n)

```

```python
#funtion to count the frequency 
def freq(a):
    d={}
    for i in a:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    print(d)
l=[1,2,2,3,4,3,3]
freq(l)

#or 
def freq(a):
    d={i:a.count(i) for i in a}
    print(d)

```

```python
#q-write a funtion who remove vowels in string.
def re_vowels(l):
    d=[]
    for i in l:
        if i not in "aeiouAEIOU":
            d.append(i.lower())
    print("".join(d))
    
re_vowels("PYTHON PROGRAMMING")

```
