1. ARGS(NON_KEYWORD ARGUMENTS)

args allows you to pass multiple positional argumants to a funtio . it collects them into a tuple.

```python
#q-1
def add_number(*args):
    total=sum(args)
    print(f"sum :{total}")
add_number(5,10,15)
```

2. KWARGS (KEYWORD ARGUMENTS):

**KWARGS ALLOWS YOU TO PASS THE KEYWORD ARGUMENT(NAME_VALUE) PAIR. IT COLLECTS THEM INTO A DICTIONARY.

```python
#Q-2
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_details(name="harshita",age=19)
```

USING ARGS AND KWARGS TOGETHER

YOU CAN USE BOTH IN SAME FUNCTION, BUT ARGS MUST COME BEFORE *KWARGS 

```python
def display_info(*args,**kwargs):
    print(args,kwargs)
    
display_info(1,2,3,name="harshita",age=19)
```

REQUIRED ARGUMENTS(MANDOTARY INPUTS)
THESE ARE ARGUMENTS THAT MUST BE PASSED TO A FUNTION:OTHERWISE ,PYTHON THROWS ERROR. THERE IS NO DEFAULT VALUE FOR THESE ARGUMENTS.

```python
def greet(name):
    print(f"hello,{name}!")
greet("Harshita")
#greet()
```

DEFAULT ARGUMENTS(OPTIONAL INPUT)

THESE ARGUMENTS HAVE A DEFAULT VALUE ASSIGNED. IF NO VALUE IS PROVIDED,PYHTON USES THE DEFALUT.

```python
def greet(name="guest"):
    print(f"{"hello",name}")
greet("harshita")
```

    RETURN KEYWORD(OUTPUT OF FUNTION)

    THE RETURN STATEMENT SENDS A VALUE BACK FROM A FUNTION. IF RETURN IS NOT USED ,THE FUNTION RETURN NONE.

```python
def add(a,b):
    return a+b
result=add(5,3)
print(result)
```

```python
#q-find max(takes multiple numbers using *args and max funtion)
def find_max(*args):
    return max(args)
print(find_max(10,15,25,14,100))
print(find_max(-1,-2,-6,0))
```

```python
#q-2 calculate the product of numbers using *args
def product(*args):
    result=1
    for n in args:
        result*=n
    return result
print(product(3,4,2))
```

```python
#q-3 write a function that takes first name and last name as keyword arguments and return ful name
def full_name(**kwargs):
    return f"{kwargs.get('first_name', 'unknown')} {kwargs.get('last_name', 'Mehta')}"
print(full_name(first_name="harshita", last_name="sharma"))
print(full_name(first_name='rahul'))
```

```python
#q-4 count the vowels of the given string using return
def count_vowel(string):
    vowels='aeiouAEIOU'
    return sum(1 for char in string if char in vowels)

print(count_vowel("hello world"))
print(count_vowel("python"))
```

```python
#q-5 create a shopping cart usinf args and kwargs
def shopping_cart(*items,**price):
    total=sum(price[item] for item in items if item in price)
    return total
print(shopping_cart("graphs","banana",graphs=50,banana=20))
print(shopping_cart("pizza","grapes",pizza=50,grapes=40))
```

```python
# funtion with required and default arguments
def greet(name,age=18):
    return f"hello {name},you are {age} years old."

print(greet("harshita"))
print(greet("harshita",19))
```

```python
#q-7 calculate discount price using **kwargs
def discount_price(**kwargs):
    price=kwargs.get("price",0)
    discount=kwargs.get("discount",0)
    return price-(price*discount/100)
print(discount_price(price=1000,discount=10))
print(discount_price(price=500,discount=20))
```

```python
# q-8 merge two lists using args
def merge_lists(*args):
    return[item for lst in args for item in lst]

print(merge_lists([1,2,3],[4,5,6]))
print(merge_lists(['a','b'],['c','d'],['e']))
```
