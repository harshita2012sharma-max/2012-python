10. Write a function that takes a number and returns its factorial (using loop).

```python
# Write your solution here
def fact(n):
    factorial=1
    while n>0:
        factorial*=n
        n=n-1
    return factorial
print(fact(5))
```

11. Write a function that takes two numbers and returns their greatest common divisor (GCD).

```python
# Write your solution here
def gcd(a,b):
    while b!=0:
        a,b=b,a%b
    return a
print(gcd(36,60))
```

12. Write a function that takes a list and returns the list without duplicates.

```python
# Write your solution here
def remove_dupe(l):
    new_list=[]
    dupe_list=[]
    for element in l:
        if element not in new_list:
            new_list.append(element)
        else:
            dupe_list.append(element)
    return new_list,dupe_list

l=[12,13,12,34,34,5,6,5,12]
new_list,dupe_list=remove_dupe(l)
print("list without duplicate",new_list)
print("duplicate only list",dupe_list)

        
```

13. Write a function that counts the vowels in a given string.

```python
# Write your solution here
def count_vowel(string):
    string=string.lower()
    only_vowel=[]
    only_consonant=[]
    for ch in string :
        if ch.isalpha():
           if ch in "aeiou":
             only_vowel.append(ch)
           else:
             only_consonant.append(ch)
    return "".join(only_consonant),"".join(only_vowel)
string="i am harshita sharma"
consonant,vowels=count_vowel(string)
print("Consonants:", consonant)
print("Vowels:", vowels)

             
```

14. Write a function that generates the Fibonacci sequence up to n terms (using loop).

```python
# Write your solution here

def fibonacci_sequence(n):
    seq = []
    a, b = 0, 1
    for i in range(n):
        seq.append(a)
        a, b = b, a + b
    return seq

print(fibonacci_sequence(5))  # Output: [0, 1, 1, 2, 3]
```

17. Write a function that checks whether a string is a palindrome.

```python
# Write your solution here
def is_palindrom(string):
    original=string
    l=[]
    for i in range(len(string)-1,-1,-1):
        l.append(string[i])
    if "".join(l)==original:
        return True
    else:
        return False
print(is_palindrom("harshita"))

    
```

18. Write a function that accepts a number and returns a list of its prime factors.

```python
# Write your solution here
def prime_factors(n):
    factors=[]
    for i in range(2,n+1):
        while n%i==0:
            factors.append(i)
            n//=i
    return factors
print(prime_factors(56))

    
```

19. Write a function that takes any number of arguments and returns their average.

```python
def make_average(*args):
    add=0
    l=len(args)
    for i in args:
        add=add+i
    return add/l
print(make_average(1,2,3,4))
```
