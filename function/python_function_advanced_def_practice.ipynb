1. Write a function that returns the common elements between two lists.

```python
# Write your solution here
def give_common(l1,l2):
    result=[]
    for i in l1:
        if i in l2:
            result.append(i)
    return result
l1=[1,2,3,4]
l2=[5,6,1,4]
print(give_common(l1,l2)) 
    
```

2. Write a function that flattens a 2D list into a 1D list.

```python
# Write your solution here
def flatten(l):
    flattened=[]
    for i in l:
        for j in i:
            flattened.append(j)
    return flattened
print(flatten([[1, 2, 3], [4, 5], [6, 7, 8]]))
```

3. Write a function that rotates a list by n positions to the right.

```python
# Write your solution here
def rotate_list(lst,n):
    if not lst:
        return []
    return lst[-n:]+lst[:-n]
print(rotate_list([4,5,6,7],2))

### means first it will print lst[-2:] means [6,7]
### then lst[:-2] means rest of the element [4,5]
# and all them 
```
