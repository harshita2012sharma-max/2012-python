```python
str1 = "Python Program"
str1 = str1.lower() 

vowels = "aeiou"
vowel_count = 0
consonant_count = 0

index = 0
while index < len(str1):
    ch = str1[index]
    if ch.isalpha():      
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1
    index += 1

print("Number of vowels:", vowel_count)
print("Number of consonants:", consonant_count)
```
